## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/astronaut.png` and show it:

```python
img=plt.imread('data/astronaut.png')

plt.imshow(img, 'gray')
plt.colorbar()
```

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img_in, size):
    result = numpy.zeros(img_in.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        #count=0
        if p[0]>=n and p[1]>=n and p[0] < img_in.shape[0]-n and p[1]< img_in.shape[1]-n:
            for q in numpy.ndindex(size, size):
                neighborhood_sum =  neighborhood_sum + img_in[p[0]-n+q[0], p[1]-n+q[1]]
                #count +=1
            result[p]= neighborhood_sum/size**2
    return result
```

**<a id="#">Task 3.1.3.</a>** Test your implementation from above by using the function `meanfilter` for the previously loaded image and show the result:

```python
img_filtered_5 = meanfilter(img,5)
plt.imshow(img_filtered_5,'gray')
```

```python
img5=plt.imread('data/astronaut_meanfilter5.png')
plt.imshow(img5, 'gray')
assert numpy.allclose(img5, img_filtered_5, atol=1/255)
```

## Non-linear filtering

**<a id="#">Task 3.2.1.</a>** Implement a median filter:

```python
def medianfilter(img_in, size):
    result = numpy.zeros(img_in.shape)
    n = (size - 1) // 2
    data=list('size')
    data.sort()
    for p in numpy.ndindex(img_in.shape):
        result[p]= data[n**2//2]
    return result
```
